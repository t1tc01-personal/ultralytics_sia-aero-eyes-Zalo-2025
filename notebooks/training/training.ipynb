## Load config

```python
import os
from pathlib import Path
from ultralytics_MB import YOLO
import torch
```

```python
# Paths
DATA_YAML = Path('/home/t1tc01-hoangphan/code/t1tc01-personal/ultralytics_sia-aero-eyes-Zalo-2025/dataset_output/data.yaml')
PROJECT_DIR = Path('/home/t1tc01-hoangphan/code/t1tc01-personal/ultralytics_sia-aero-eyes-Zalo-2025/runs/detect')
```

```python
# Model configuration
MODEL_CONFIG = 'yolov9es.yaml'  # Siamese YOLOv9e-s (ES version)
```

```python
# Training hyperparameters
TRAINING_CONFIG = {
    # Siamese settings
    'channels': 6,              # 3 channels reference + 3 channels query
    'bands': [0, 1, 2, 3, 4, 5], # All 6 bands
    'dual_stream': True,         # Enable siamese dual-stream
    
    # Training
    'epochs': 100,
    'batch': 8,                 # Adjust based on GPU memory (8GB GPU: 8-16, 16GB+: 32)
    'imgsz': 640,                # Match v·ªõi target_size trong conversion
    'lr0': 0.001,                # Initial learning rate
    
    # Device
    'device': 0 if torch.cuda.is_available() else 'cpu',
    
    # Optimization
    'optimizer': 'auto',
    'weight_decay': 0.0005,
    'momentum': 0.937,
    
    # Learning rate schedule
    'cos_lr': False,
    'warmup_epochs': 3.0,
    'warmup_momentum': 0.8,
    'warmup_bias_lr': 0.1,
    
    # Loss weights
    'box': 7.5,
    'cls': 0.5,
    'dfl': 1.5,
    
    # Data augmentation
    'hsv_h': 0.015,
    'hsv_s': 0.7,
    'hsv_v': 0.4,
    'degrees': 0.0,
    'translate': 0.1,
    'scale': 0.5,
    'shear': 0.0,
    'perspective': 0.0,
    'flipud': 0.0,
    'fliplr': 0.5,
    'mosaic': 1.0,
    'mixup': 0.1,
    'copy_paste': 0.0,
    
    # Validation
    'val': True,
    'split': 'val',
    
    # Saving
    'save': True,
    'save_period': 10,
    
    # Other
    'workers': 8,
    'project': str(PROJECT_DIR),
    'name': 'siamese_yolov9e_zalo',
    'exist_ok': True,
    'pretrained': True,
    'verbose': True,
    'seed': 0,
    'deterministic': True,
    'amp': True,
    'fraction': 1.0,
    'patience': 50,
    'close_mosaic': 10,
}

# Fusion method (optional - uncomment to change)
# os.environ['FUSION_METHOD'] = 'diff'        # Default: difference
# os.environ['FUSION_METHOD'] = 'attention'   # Attention fusion
# os.environ['FUSION_METHOD'] = 'add'         # Addition
# os.environ['FUSION_METHOD'] = 'multiply'    # Multiplication
```

```python
print("=" * 70)
print("üìã TRAINING CONFIGURATION")
print("=" * 70)
print(f"Dataset: {DATA_YAML}")
print(f"Model: {MODEL_CONFIG}")
print(f"Device: {TRAINING_CONFIG['device']}")
print(f"Epochs: {TRAINING_CONFIG['epochs']}")
print(f"Batch size: {TRAINING_CONFIG['batch']}")
print(f"Image size: {TRAINING_CONFIG['imgsz']}")
print(f"Channels: {TRAINING_CONFIG['channels']}")
print(f"Dual stream: {TRAINING_CONFIG['dual_stream']}")
print(f"Learning rate: {TRAINING_CONFIG['lr0']}")
print(f"Fusion method: {os.environ.get('FUSION_METHOD', 'diff')} (default)")
print("=" * 70)
```

## Load model

```python
# ============================================================================
# LOAD MODEL
# ============================================================================

print("=" * 70)
print("üöÄ LOADING SIAMESE YOLOv9e MODEL")
print("=" * 70)

# Load siamese model (ES version)
model = YOLO(MODEL_CONFIG, task='detect')

print(f"‚úÖ Model loaded: {MODEL_CONFIG}")
print(f"‚úÖ Task: Object Detection")
print(f"‚úÖ Siamese: {TRAINING_CONFIG['dual_stream']}")
```

```python
# Check if dataset exists
if not DATA_YAML.exists():
    raise FileNotFoundError(f"‚ùå Dataset not found: {DATA_YAML}")

print(f"‚úÖ Dataset found: {DATA_YAML}")
```

```python
# Check GPU
if torch.cuda.is_available():
    print(f"‚úÖ GPU available: {torch.cuda.get_device_name(0)}")
    gpu_memory = torch.cuda.get_device_properties(0).total_memory / 1e9
    print(f"   GPU Memory: {gpu_memory:.1f} GB")
    
    # Recommend batch size
    if gpu_memory < 8:
        print(f"   ‚ö†Ô∏è  Small GPU - consider reducing batch size to 8")
    elif gpu_memory < 16:
        print(f"   ‚úÖ Medium GPU - batch size {TRAINING_CONFIG['batch']} should work")
    else:
        print(f"   ‚úÖ Large GPU - can increase batch size to 32+")
else:
    print("‚ö†Ô∏è  No GPU available, using CPU (training will be very slow)")

print("=" * 70)
```

## Start training

```python
print("\n" + "=" * 70)
print("üéØ STARTING TRAINING")
print("=" * 70)
print("Training will start now...")
print("Monitor progress in the output below and in the results directory")
print("=" * 70)

try:
    # Start training
    results = model.train(
        data=str(DATA_YAML),
        **TRAINING_CONFIG
    )
    
    print("\n" + "=" * 70)
    print("‚úÖ TRAINING COMPLETE!")
    print("=" * 70)
    print(f"üìÅ Results directory: {results.save_dir}")
    print(f"üèÜ Best model: {results.save_dir}/weights/best.pt")
    print(f"üìù Last model: {results.save_dir}/weights/last.pt")
    
    # Print final metrics if available
    if hasattr(results, 'results_dict'):
        print(f"\nüìä Final Metrics:")
        for key, value in results.results_dict.items():
            if isinstance(value, (int, float)):
                print(f"   {key}: {value:.4f}")
    
    print("=" * 70)
    
except KeyboardInterrupt:
    print("\n‚ö†Ô∏è  Training interrupted by user")
    print("üí° You can resume training later using:")
    print(f"   model = YOLO('runs/detect/siamese_yolov9e_zalo/weights/last.pt')")
    print(f"   model.train(resume=True)")
except Exception as e:
    print(f"\n‚ùå Training failed: {e}")
    import traceback
    traceback.print_exc()
    raise
```

## Evaluate

```python
# ============================================================================
# EVALUATE BEST MODEL (Optional)
# ============================================================================

if 'results' in locals() and hasattr(results, 'save_dir'):
    best_model_path = Path(results.save_dir) / 'weights' / 'best.pt'
    
    if best_model_path.exists():
        print("=" * 70)
        print("üìä EVALUATING BEST MODEL")
        print("=" * 70)
        
        # Load best model
        best_model = YOLO(str(best_model_path))
        
        # Validate
        print("Running validation on best model...")
        metrics = best_model.val(
            data=str(DATA_YAML),
            imgsz=TRAINING_CONFIG['imgsz'],
            batch=TRAINING_CONFIG['batch'],
            device=TRAINING_CONFIG['device']
        )
        
        print("\nüìä Validation Results:")
        if hasattr(metrics, 'results_dict'):
            for key, value in metrics.results_dict.items():
                if isinstance(value, (int, float)):
                    print(f"   {key}: {value:.4f}")
        
        print("=" * 70)
    else:
        print("‚ö†Ô∏è  Best model not found - training may not have completed")
else:
    print("‚ö†Ô∏è  No training results found - run training cell first")
```
